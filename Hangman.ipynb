```python
import random
import string

#list created to get the computer to choose automatically a word from

guessing_list=["cat","dog","sloth","rabbit","meerkat"]

print("Welcome to Hangman!")
print("The computer will randomly select a word from a list and you have to guess the word!")
print("You will start with seven lives, the length of HANGMAN!")
print("If the letter is not in the word, you will start losing lives - and you untilmately lose when the man is hanged and zero lives are left")
print("Hopefully you will guess the letters of the word before you KILL THE MAN and WIN. GOOD LUCK!")


#function to randomly choose a word from the list

def get_random_word(initial_list):
    word = random.choice(initial_list) 
    return word.lower()

#main hangman game function to call for repetions 

def hangman():
    attempts = 7 #hangman length
    word = get_random_word(guessing_list) #calls the previous function that chooses the word
    word_letters = set(word)  #stores all the letters in the word; essential to break the while loop and set the winner
    all_letters = set(string.ascii_lowercase) #list of the lowercase letters from a to z - string operations, to avoid writing the alphabet
    guessed_letters = set()  # empty set to store the letter guessed already: set to avoid repetitions
    

    # getting user input
    #len=0 defines winning, attempts=0 defines losing. games continuous until one of them happens
    while len(word_letters) > 0 and attempts > 0:
        
        # to print letters used and attempts left
        print(f'You have {attempts} attempts left and the following letters were used already {" ".join(guessed_letters)}')
        
        # list comprehension to show letter if in guessed letters,otherwise to show ? to give the player the info about letters missing
        word_list = [letter if letter in guessed_letters else '?' for letter in word]
        print(f'Current word: {" ".join(word_list)}') #just to add a space, could call just word_list
        
        #asking user input
        player_letter = input("Please write your guess: ").lower() 
        
        
        #if input is valid, player letter is in all_letters
        #all_letters excluding the guessed_letters, so it doesn´t consider the letters guessed already, it works but not sure if sintax is appropriate 
        if player_letter in all_letters-guessed_letters: 
            guessed_letters.add(player_letter) #add the new letter to the used letters set, set useful for no repetitions
            
            #scenario 1: if the input is right and in the word
            if player_letter in word_letters: #if user input is a letter in word (word letters is the set of word)
                word_letters.remove(player_letter) #then its going to be removed from word_letters and it´s basically to define the winner (while loop breaks when all removed), therefore no more words to guess
                print(f'GREAT GUESS!! The letter {player_letter} is part of the word')
                
            #scenario 2:Input is right but not in the word
            else:
                attempts-= 1  # attempts decreasing due to failure, important for the while loop, to define losing and end of the game
                print(f'NOT GOOD ENOUGH!! The letter {player_letter} is not part of the word')
                position=6-attempts
                hangman=["H","HA","HAN","HANG","HANGM","HANGMA","HANGMAN"] #showing hangman, while HANGMAN, player loses
                print(f'Adding a letter to HANGMAN...and HANGMAN means DEATH to the nice man :(...  {hangman[position]}')
                
        #scenario 3: input is right but letter already guessed
        elif player_letter in guessed_letters: #nothing happens, just a reminder that the letter was used already
            print("ATTENTION!!The letter was used already. Guess another letter please")

        #scenario 4: input is wrong (typo etc) 
        else: #if its invalid, back to the cycle to ask input
            print("ATTENTION!!That is not a valid letter")

    # gets here when len(word_letters) == 0 OR when lives == 0, the deal breakers for winning or losing 
    if attempts == 0:
        print(f'YOU LOST THE GAME! :( BETTER LUCK NEXT TIME. THE WORD WAS {word.upper()}')
    else:
    #elif word==used_letters:
        print(f'YOU WON, CHAAAMP!! :D THE WORD WAS {word.upper()}')

#additional function to define if the player wants to play another time or not
        
def play_again():
        play=input("Do you want to play one more round? :) write y or n for yes or no!").lower()
        while (play!="y") and (play !="n"):
            play=input("Wrong input. pelase write only y or n").lower()
        else:
            if play=="y":
                return hangman()
            elif play=="n":
                print("THANK YOU FOR PLAYING, SEE YOU SOON!!")

#calling functions

hangman()
play_again()
```
