```python
import random
import string

#list created to get the computer to choose automatically a word from

guessing_list=["cat","dog","sloth",]

print("Welcome to Hangman!")
print("The computer will randomly select a word from a list and you have to guess the word!")
print("You will start with seven lives, the length of HANGMAN!")
print("If the letter is not in the word, you will start losing lives - and you untilmately lose when the man is hanged and zero lives are left")
print("Hopefully you will guess the letters of the word before you KILL THE MAN and WIN. GOOD LUCK!")


#function to randomly choose a word from the list

def get_random_word(initial_list):
    word = random.choice(initial_list) 
    return word.lower()

#main hangman game function to call for repetions 

def hangman():
    attempts = 7 #hangman len
    word = get_random_word(guessing_list) #calls the previous function that chooses the word
    word_letters = set(word)  # letters in the word, saves all the letters
    all_letters = set(string.ascii_lowercase) #list of the lowercase letters from a to z - string operations, to avoid write the alphabet
    guessed_letters = set()  # empty set to store the letter guessed already, to show the player the options left
    

    # getting user input
    #len defines winning, attempts define losing.running until one of them happens
    while len(word_letters) > 0 and attempts > 0:
        
        # to print letters used and attempts left
        print(f'You have {attempts} attempts left and the following letters were used already {" ".join(guessed_letters)}')
        
        # list comprehension to show letter if in guessed letters,otherwise to show ? to give the player the info about letters missing
        word_list = [letter if letter in guessed_letters else '?' for letter in word]
        print(f'Current word: {" ".join(word_list)}') #just to add the info from word_list #personal note:tried calling only word_list and it works but it gives a list obviously
        
        #asking user input
        user_letter = input("Please write your guess: ").lower() 
        
        #if input is valid and hasnÂ´t been used yet
        if user_letter in all_letters-guessed_letters: #if the input is a valid character that hasnt been used yet
            guessed_letters.add(user_letter) #add it to used letters list if it hasnt been used yet
            #The set add() method adds a given element to a set. If the element is already present, it doesn't add any element.
            
            #scenario 1: if the input is right and in the word
            if user_letter in word_letters: #if user input is in the word
                #saw in the internet, not understanding well
                word_letters.remove(user_letter) #then its going to be removed from word_letters
                print("GOOD GUESS!!")
                #print("good guess! the letter", user_letter, "is part of the word")
            #scenario 2:Input is right but not in the word
            else:
                attempts-= 1  # attempts decreasing due to failure, when letter is not part of the word and player has less rounds to play
                print(f'NOT GOOD ENOUGH!! The letter {user_letter} is not part of the word')
                position=6-attempts
                hangman=["H","HA","HAN","HANG","HANGM","HANGMA","HANGMAN"]
                print(f'Adding a letter to HANGMAN...and HANGMAN means DEATH to the nice man :(...  {hangman[position]}')
                
        #scenario 3: input is right but letter already guessed
        elif user_letter in guessed_letters: #nothing happens, just a reminder that the letter was used already
            print("ATTENTION!!The letter was used already. Guess another letter please")

        #scenario 4: input is wrong (typo etc) 
        else: #if its invalid, back to the cycle to ask input
            print("ATTENTION!!That is not a valid letter")

    # gets here when len(word_letters) == 0 OR when lives == 0, the deal breakers for winning or losing 
    if attempts == 0:
        print(f'You LOST the game! :( Better luck next time. The word was {word}')
    else:
    #elif word==used_letters:
        print(f'YOU WON, GREAT GUESS!! :D THE WORD WAS {word}')


def play_again():
        play=input("Do you want to play one more round? :) write y or n for yes or no!").lower()
        while (play!="y") and (play !="n"):
            play=input("Wrong input. pelase write only y or n").lower()
        else:
            if play=="y":
                return hangman()
            elif play=="n":
                print("GAME OVER")
                       
hangman()
play_again()
```
